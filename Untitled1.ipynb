```python
import os
# os.environ['DEVICE_ID'] = '6'
import csv
import numpy as np
import mindspore as ms
from mindspore import nn
from mindspore import context
import mindspore.dataset as ds
from mindspore.train.callback import LossMonitor
from mindspore.common.api import ms_function
from mindspore.ops import operations as P
context.set_context(mode=context.GRAPH_MODE, device_target="CPU")#导入必要的软件包
```

```python
import pandas as pd
iris = pd.read_csv("D:\ipynb\iris.csv",header=None)
print(iris[:100])#读取数据并查看部分数据，读取前两类的数据，因为第一类与后两类都可以线性分离，二后两类无法线性分类
```

```python
iris_data=iris[:100]
iris_data.iloc[:,4].replace("Iris-setosa",0,inplace=True)
iris_data.iloc[:,4].replace("Iris-versicolor",1,inplace=True)#将两类的种类分别映射为0，1
print(iris_data)#展示数据
```

```python
import matplotlib.pyplot as plt
x_1=iris_data.iloc[:50,0]#将数据分类
y_1=iris_data.iloc[:50,1]
x_2=iris_data.iloc[50:,0]
y_2=iris_data.iloc[50:,1]

plt.figure(figsize=(8, 6))
plt.clf()
plt.scatter(x_1,y_1,label='Iris-setosa',color='blue')
plt.scatter(x_2,y_2,label='Iris-virginica',color='orange')

plt.xlabel('Sepal length')#x轴和y轴
plt.ylabel('Sepal width')
plt.xticks(np.arange(4.5,7.5,0.5))
plt.yticks(np.arange(2.0,5,0.5))
plt.legend()
plt.show()#画图
```

```python
from sklearn.model_selection import train_test_split
iris_data_train, iris_data_test,\
iris_target_train, iris_target_test = \
train_test_split(iris_data.iloc[:,:4], iris_data.iloc[:,4],
    test_size=0.2, random_state=42)#将训练集和测试集分类
```

```python
from mindspore.dataset import GeneratorDataset

class DatasetGenerator:
    def __init__(self,x,y):
        self.data = x
        self.label = y

    def __getitem__(self, index):
        return self.data[index], self.label[index]

    def __len__(self):
        return len(self.data)   
iris_data_train = np.array(iris_data_train)#将类别转化为array
iris_data_test = np.array(iris_data_test)
iris_target_train = np.array(iris_target_train)
iris_target_test = np.array(iris_target_test)
dataset_generator = DatasetGenerator(iris_data_train,iris_target_train)
dataset = ds.GeneratorDataset(dataset_generator, ["data", "label"], shuffle=False)#利用GeneratorDataset接口将array转化为dataset
for data in dataset.create_dict_iterator():
    print('{}'.format(data["data"]), '{}'.format(data["label"]))#迭代输出dataset的数据
```

```python
import numpy as np #可视化逻辑回归函数sigmoid
from matplotlib import pyplot as plt

def sigmoid(x):
   return 1/(1+np.exp(-x))

x = np.linspace(-10,10,num=100) #生成数据
y = sigmoid(x)
plt.subplot(1,2,1)
plt.plot(x,y)
plt.show()
```

```python
net = nn.Dense(4,1)#输入时（4，1）的向量
data = ms.Tensor(iris_data_train,ms.float32)#转化为张量
z = net(data)#利用dense转化为一个元素的向量
iris_target_train = iris_target_train.reshape(-1,1)#重新确定大小
iris_target_train = ms.Tensor(iris_target_train,ms.float32)#转化为张量
loss = P.SigmoidCrossEntropyWithLogits()#使用SigmoidCrossEntropyWithLogits 算 子
loss_sum = sum(loss(z,iris_target_train))#算总和
loss_ave = loss_sum/(len(iris_target_train))#得到平均损失值
print(loss_ave)
```

```python
from sklearn.model_selection import train_test_split
loss = []
max_w=0
max_b=0
acc=0
for i in range(200):#按照上述步骤进行迭代训练
    iris_data_train, iris_data_test,\
    iris_target_train, iris_target_test = \
    train_test_split(iris_data.iloc[:,:4], iris_data.iloc[:,4],
        test_size=0.2, random_state=np.random.randint(1,10))
    from mindspore.dataset import GeneratorDataset
    w = np.random.randint(-50,50)#随机生成权重和偏置
    b = np.random.randint(-30,30)
    class DatasetGenerator:
        def __init__(self,x,y):
            self.data = x
            self.label = y

        def __getitem__(self, index):
            return self.data[index], self.label[index]

        def __len__(self):
            return len(self.data)   
    iris_data_train = np.array(iris_data_train)
    iris_data_test = np.array(iris_data_test)
    iris_target_train = np.array(iris_target_train)
    iris_target_test = np.array(iris_target_test)
    dataset_generator = DatasetGenerator(iris_data_train,iris_target_train)
    dataset = ds.GeneratorDataset(dataset_generator, ["data", "label"], shuffle=False)
    net = nn.Dense(4,1,weight_init=w, bias_init=b)
    data = ms.Tensor(iris_data_train,ms.float32)
    z = net(data)
    iris_target_train = iris_target_train.reshape(-1,1)
    iris_target_train = ms.Tensor(iris_target_train,ms.float32)
    sigmoid=P.Sigmoid()
    predict=sigmoid(net(ms.Tensor(iris_data_test,ms.float32)))
    predict_target = np.where(predict >= 0.5,1,0)#只要概率大于百分之50就认为是种类1
    Loss = P.SigmoidCrossEntropyWithLogits()
    loss_sum = sum(Loss(z,iris_target_train))
    loss_ave = loss_sum/(len(iris_target_train))
    loss_ave = ms.Tensor.asnumpy(loss_ave)
    loss.append(loss_ave)
    print('第',i+1,'次精度：',(predict_target==iris_target_test).mean())#算出精度
    print('第',i+1,'次的交叉熵损失值：',loss_ave)#算出交叉熵损失值
    print('权重:',w,'偏置：',b)
    if(acc<(predict_target==iris_target_test).mean()):
        acc= (predict_target==iris_target_test).mean()
        max_w=w
        max_b=b
print('精度最大为：',acc,' ','权重和偏置分别为：',max_w,' ',max_b)
```
